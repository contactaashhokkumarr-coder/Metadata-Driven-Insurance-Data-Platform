```python
%sql
ALTER TABLE insurance_metadata.bronze_config
ADD COLUMNS
(
    gold_table STRING,
    load_strategy STRING,
    active STRING
);
```

```python
%sql
ALTER TABLE insurance_metadata.bronze_config
ADD COLUMNS
(
    history_table STRING,
    compare_columns STRING,
    scd_enabled STRING
);
```

```python
%sql
UPDATE insurance_metadata.bronze_config
SET
    history_table = 'dim_customer_history',
    compare_columns = 'first_name,last_name,email,phone,country,city',
    scd_enabled = 'Y'
WHERE lower(table_name) = 'customer';
```
